### Create a scatter plot that shows the relationship between 'sugarpercent' (on the horizontal x-axis) and 'winpercent' (on the vertical y-axis).

### Given the dataset above, solve the following:
A) Create a DataFrame that includes information about candy bars, including their names, sugar percentage, and winning percentage.

B) Create a figure for the scatter plot with a size of 10 by 6 inches.

C) Plot a scatter plot using the 'sugarpercent' column for the x-axis and the 'winpercent' column for the y-axis. Use the color 'skyblue' for the points, set the size of the markers to 200, and apply a black edge color.

D) Titles and Labels: Add a title to the scatter plot that reads "Sugar Percentage vs Winning Percentage of Candy Bars". Label the x-axis as "Sugar Percentage (%)" and the y-axis as "Winning Percentage (%)".

E) Enable the grid in the scatter plot for better readability of data points.

F) Finally, display the plot using the appropriate function.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.scatter(df['sugarpercent'], df['winpercent'], color='skyblue', s=200, edgecolor='black')

# Add titles and labels
plt.title('Sugar Percentage vs Winning Percentage of Candy Bars')
plt.xlabel('Sugar Percentage (%)')
plt.ylabel('Winning Percentage (%)')

# Annotate points with competitor names (Additional step - for your infformation)
for i in range(len(df)):
    plt.annotate(df['competitorname'][i], (df['sugarpercent'][i], df['winpercent'][i]),
                 textcoords="offset points", xytext=(0, 10), ha='center')

# Show grid and plot
plt.grid(True)
plt.show()
```

### In the cell below, you should create a scatter plot to show the relationship between 'pricepercent' (on the horizontal x-axis) and 'winpercent' (on the vertical y-axis). Use the 'chocolate' column to color-code the points.


A) Create a DataFrame that includes information about candy bars, including their names, price percentages, winning percentages, and whether they contain chocolate.

B) Create a figure for the scatter plot with a size of 10 by 6 inches.

C) Create a color array where points are colored blue if the candy bar contains chocolate (indicated by 'Yes' in the 'chocolate' column) and orange if it does not.

D) Plot a scatter plot using the 'pricepercent' column for the x-axis and the 'winpercent' column for the y-axis. Use the color array created in the previous step for the points, set the size of the markers to 200, and apply a black edge color.

E) Titles and Labels: Add a title to the scatter plot that reads "Price Percentage vs Winning Percentage of Candy Bars". Label the x-axis as "Price Percentage (%)" and the y-axis as "Winning Percentage (%)".

F) Enable the grid in the scatter plot for better readability of data points.

G) Finally, display the plot using the appropriate function.

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))  # Set the figure size

# Color-coding based on the 'chocolate' column
colors = np.where(df['chocolate'] == 'Yes', 'blue', 'orange')

plt.scatter(df['pricepercent'], df['winpercent'], color=colors, s=200, edgecolor='black')

# Add titles and labels
plt.title('Price Percentage vs Winning Percentage of Candy Bars')
plt.xlabel('Price Percentage (%)')
plt.ylabel('Winning Percentage (%)')

# Annotate points with competitor names
for i in range(len(df)):
    plt.annotate(df['competitorname'][i], (df['pricepercent'][i], df['winpercent'][i]),
                 textcoords="offset points", xytext=(0, 10), ha='center')

# Show grid and plot
plt.grid(True)
plt.show()
```
