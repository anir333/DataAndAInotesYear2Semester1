### Precision
1- Precision measures the proportion of positive predictions that are actually correct. It answers the question

#### Questions
1- Of all the instances the model predicted as positive, how many were truly positive?

2- What proportion of positive identifications was actually correct?	


#### Formula
1- P= Number of total positive / number of the pred positive

2- P= False Positives (FP) / False Positives (FP)+True Negatives (TN)


### Recall
Recall (also called Sensitivity or True Positive Rate) measures the proportion of actual positive instances that the model correctly identified.

#### Questions
1- Out of all the actual positives, how many did the model correctly predict?

2- What proportion of actual positives was classified correctly?


#### Formula
1- R= Number of TP / number of actual total positive

2- R= True Positives (TP) / True Positives (TP)+False Negatives (FN)

##### Recall expresses which instances are relevant in a data set. It is important to examine both the Precision AND Recall when evaluating a model because they often have an inverse relationship. When precision increases, recall tends to decrease and vice versa.

### Accuracy
Accuracy is the most intuitive metric: it measures the proportion of total correct predictions out of all predictions made. While accuracy is useful, it can be misleading in imbalanced datasets (where one class dominates).

#### Questions
How many did we classify correctly?


#### Formula
	
1- A= number of true positive + True negatives / Total observations

2- A= TP+TN/ TP+FP+TN+FN	

### F1-Score
is a function of precision and recall. It is used to find the correct balance between the two metrics.

#### Questions
It determines how many instances your model classifies correctly without missing a significant number of instances.


#### Formula
f1= 2 ((precision*Recall)/(Precsion+Recall))


##### Having an imbalance between precision and recall, such as a high precision and low recall, can give you an extremely accurate model, but classifies difficult data incorrectly. We want the F1 Score to be as high as possible for the best performance of our model


#### Specificity 
is also known as True Negative Rate.

It is a measure of negative examples labeled as negative by a classifier. There should be high specificity. For instance, proportion of emails that are non-spam among all non-spam emails.
