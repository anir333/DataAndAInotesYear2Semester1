## Overfitting
A common problem in machine learning and deep learning algorithms:

- The model performs well on the training datasets.
- Perform poorly on the test dataset

### Reasons: 

- The complexity of the model

  - The model captures noise
  - Random variations in the training datasets (outliers, or any idiosyncratic patterns) 
  - instead of detecting underlying patterns and relationships from the dataset

- No Generlatzation on unseen data

- Low accuracy (random fluctuation in the training curves)


### ML (occurs)
- Model has too many parameters w.r.t the training dataset
- A large number of epochs (training for a long time)

  - The model memorizes the training dataset, instead of training patterns 
  - This leads to poor generalization performance 

### DL (occurs)
- Model has too many layers (this means the model is too deep)
- This leads to the production of a huge amount of parameters 

  - This is a very well-known issue in  DL 
  - Because of this huge number of parameters
  - It is more prone to overfitting if it is not regularized very well 

### How can you detect overfitting 
- Validation set:

  - Split the dataset for training and validation set
  - Validation: evaluate the model performance using unseen dataset

        - If the performance -> decrease (validation set)
        - the performance -> increase (training set)
    - A big sign of overfitting 
  
- Learning curves:

      - PLot the performance of the model (learning curves - valid + training)

      -  If the training curve keeps increasing

      - The valid start decreasing

  - A big sign of overfitting (This might mean a complex model and little data) 

- Cross validations (split the dataset for multi-folds)

      - Train the model on different combinations of folds

      - Evaluate the model performance using the rest of the folds

      - If the performance of the validation is worse (big gap)
   - A big sign of overfitting 


### Prevent overfitting

- Early stopping
  - Monitoring the performance (validation and the training)
  - Stop the training when the validation performance starts decreasing \ there is no improvement anymore

    
- Regularization

  - Adding a penalty term to the loss function
  - This helps the model (to have smaller weights) and (be less complex)
  - Check the validation performance (should be improved)

      - Lasso regression ($L_1$ regularization):
   
             - Another very common type of regularization is known as lasso, and involves penalizing the sum of absolute values (1-norms) of regression coefficients:
   
             - lasso regression tends to favor *sparse models* where possible: that is, it preferentially sets model coefficients to exactly zero.
   
             - Add the penalty to the absolute values of the weights 
   
        $$
P = \alpha\sum_{n=1}^N |\theta_n|
$$

      - Ridge regression ($L_2$ Regularization)
   
             - Perhaps the most common form of regularization is known as *ridge regression* or $L_2$ *regularization*
   
             - This proceeds by penalizing the sum of squares (2-norms) of the model coefficients
   
             - Add the penalty to the absolute values of the weights
   
             - $\alpha$ is a free parameter that controls the strength of the penalty.

$$
P = \alpha\sum_{n=1}^N \theta_n^2
$$

   
- Dropout
    - Randomly drop some units in the neural network during the training
    - Prevent overfitting by forcing the network to learn more robust features (useful features)
   
- Data augmentation

      - Create new datasets by applying random transformations

      - Increase the amount of the training datasets



- Simplifying your model

      - If the model is complex then it might be overfitting

      - You should make your model simple by reducing its complexity

          - Reduce the number of hidden layers (NN)

          - Reduce the number of units in each layer

          - Number of features in the model 

#### In overfitting: 
- training acc greater than the validation
    - The model memorizing noise patterns
      
- The goal is to have both high acc

      - Model generalizes well

      - So: it is really important to monitor the training process

#### Overfitting, means high variance
    - Our goal is to reduce the variance without affecting the bias

## Underfitting
A common problem in machine learning and deep learning algorithms:

- The model does not perform well on both training and validation datasets.
  

### Underfitting
- The model is too simple to capture the underlying patters
- The model is not able to fit the dataset well
- Perform poorly on both training and validation datasets
- This means, the model does not learn enough from the data

### Detect underfitting
1- Low training and validation accuracies
    
    - Model is too simple
    - Not able to generalize on new data points

2- Learning curves
    
    -Model not able to improve on the training and validation datasets
    

3- Error analysis

    - Error (large) on both datasets
    - The model is not able to capture the complexity of the dataset

### Addressing the underfitting 
1- Increase the model complexity

    - Increase the number of layers\ neurons in NN 
    - Increase the number of parameters (linear model)
2- Add more features to the model

    - Feature selection (feature importance using RF)
3- Collect more data

    - Lack of data
    - Data augmentation \ collect new data
4- Reduce regularizations

    - Reduce the strength of the regularizations
    - Reduce the number of parameters of the regularizations
5- Tune-hyper parameters

    - Learning rate
    - Regularizations arguments

## Missing values


1. Less than 5% Missing Data:
If less than 5% of your data is missing, it's usually safe to remove the rows or columns with missing values without significantly affecting the analysis.

2. 5% to 20% Missing Data:
In this range, you need to be more cautious. Dropping rows or columns might still be acceptable in certain cases, but you should consider other methods, such as imputation (filling in missing values).


3. More than 20% Missing Data:
If more than 20% of the data is missing, simply removing those rows or columns may cause significant loss of information and can introduce bias or distort the analysis.
